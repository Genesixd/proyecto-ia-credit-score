1.5. Limpieza básica: valores raros ⇒ NaN

En este dataset hay valores “basura” tipo "_", "__10000__", "NA", "NM", etc. La idea es tratarlos como faltantes.

```python
import numpy as np

# Lista de tokens que consideraremos como "faltantes"
rare_tokens = ["_", "__", "NA", "NM", "Not Specified", "__10000__", "!", "!@9#%8", "-"]

df = df.replace(rare_tokens, np.nan)

#Si luego ves otros valores raros en tus columnas, los agregas a esa lista. <------------ OJO
```

Manejo de valores faltantes

```python
#Imputar numéricas con la mediana
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

for col in num_cols:
    med = df[col].median()
    df[col] = df[col].fillna(med)
```

```python
#Imputar categóricas con la moda
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    mode = df[col].mode(dropna=True)
    if len(mode) > 0:
        df[col] = df[col].fillna(mode[0])
    else:
        df[col] = df[col].fillna("Unknown")
```

```python
#Verificar que ya no queden NaN
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
#Guardar una versión limpia para usar luego
df.to_csv("credit_score_clean.csv", index=False)
```

# 2. Objetivo general de la investigación

## 2.1 Planteamiento del problema

En este proyecto se aborda el problema de **clasificación del puntaje crediticio** de clientes de una entidad financiera.  
A partir de información demográfica, financiera y de comportamiento de pago, se busca predecir la categoría de **`Credit_Score`** de cada cliente (`Poor`, `Standard` o `Good`).

## 2.2 Objetivo general

**Desarrollar y evaluar modelos de aprendizaje automático capaces de clasificar el puntaje crediticio (`Credit_Score`) de los clientes, utilizando como entrada sus características financieras y de comportamiento de pago.**

## 2.3 Objetivos específicos

- Analizar y preprocesar el dataset `train.csv`, manejando valores faltantes, valores atípicos y variables categóricas.
- Entrenar y comparar distintos clasificadores supervisados para la predicción de `Credit_Score`.
- Evaluar el rendimiento de los modelos mediante métricas como accuracy, precision, recall y F1-score, utilizando diferentes particiones de los datos.
- Identificar qué variables tienen mayor influencia en la clasificación del puntaje crediticio.

## 2.4 Justificación del estudio

En el contexto financiero actual, la correcta evaluación del riesgo crediticio es un requisito fundamental para la sostenibilidad de las entidades financieras. Cada decisión de aprobar o rechazar un crédito implica un equilibrio entre rentabilidad y riesgo: otorgar préstamos a clientes con alto riesgo de impago puede generar pérdidas significativas, mientras que rechazar injustificadamente a clientes solventes limita las oportunidades de negocio. En este escenario, contar con herramientas objetivas y sistemáticas para clasificar el puntaje crediticio se vuelve esencial.

Tradicionalmente, la evaluación del riesgo se ha realizado de manera manual, apoyándose en la experiencia de los analistas de crédito y en reglas de negocio predefinidas. Este enfoque presenta varias limitaciones: es lento, no escala bien cuando el volumen de clientes crece, está expuesto a sesgos humanos y puede desaprovechar la gran cantidad de datos históricos disponibles. El uso de técnicas de aprendizaje automático permite automatizar parte de este proceso, explotando patrones complejos en los datos que no siempre son evidentes a simple vista.

El dataset utilizado en este proyecto integra información demográfica, financiera y de comportamiento de pago de los clientes, lo que permite modelar el riesgo crediticio de forma más integral. Desarrollar un modelo capaz de predecir la categoría de `Credit_Score` (Poor, Standard o Good) a partir de estas variables no solo reduce el esfuerzo manual, sino que también puede contribuir a decisiones más consistentes, transparentes y basadas en evidencia. Además, desde el punto de vista académico, este estudio permite aplicar de manera práctica todo el flujo de un proyecto de Inteligencia Artificial: desde la limpieza y preprocesamiento de datos hasta la comparación de modelos, el análisis de desempeño y la interpretación de resultados, fortaleciendo las competencias técnicas necesarias para abordar problemas reales en el ámbito financiero.
